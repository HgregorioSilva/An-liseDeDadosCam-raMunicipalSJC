
# Nome e salário bruto ordenado

# 10 Maiores salários Câmara da prefeitura de São José dos Campos-SP

```python
Maiores_10_salarios = df[['Nome do Servidor','Salário Bruto']].set_index('Nome do Servidor').sort_values('Salário Bruto', ascending=False).head(10)
Maiores_10_salarios.plot(kind='barh', figsize=(7,3), grid=False, color='blue', legend=False)

plt.title('10 maiores salários dos Servidores da Câmara de SJC em maio/2019')
plt.xlabel('Total em reais')
plt.show()
```

# 10 maiores salários por cargos Câmara de SJC em maio/2019

```python
CargosMaioresSalarios = df[['Cargo','Salário Bruto']].sort_values('Salário Bruto', ascending=False).head(10).set_index('Cargo')
CargosMaioresSalarios.plot(kind='barh', figsize=(7,3), grid=False, color='black', legend=False)

plt.title('10 maiores salários por cargos Câmara de SJC em maio/2019')
plt.xlabel('Total em reais')
plt.show()
```
